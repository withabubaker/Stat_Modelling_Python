Build a regression model.

```python
import statsmodels.api as sm
import pandas as pd
df = pd.read_csv('join.csv', index_col=0)
df.head()
```

Provide model output and an interpretation of the results. 

```python
X = df[['street_c','postCode_c','rating','category_c']] 
y = df['price_c'] 
X = sm.add_constant(X) 
est = sm.OLS(y, X).fit() 
est.summary()
```
